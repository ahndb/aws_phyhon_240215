#### `Axes` 객체와 subplot  
하나의 Figure 객체 안에 여러개의 Axes 객체를 생성하여 행렬 형태의 subplot을 생성할 수 있음  
이때 `subplot()` 메서드를 사용하여 명시적으로 Axes 객체를 생성할 수 있음  
그러나 pyplot의 `plot()` 메서드를 바로 사용해도 Axes 객체를 생성할 수 있음

```python
x1 = np.linspace(0, 5)
x2 = np.linspace(0, 2)
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

plt.subplot(2, 1, 1)
plt.plot(x1, y1, 'yo-')
plt.title('2행 1열 서브 플롯 첫번째')

plt.subplot(2, 1, 2)
plt.plot(x2, y2, 'r.-')
plt.title('2행 1열 서브 플롯 두번째')

plt.tight_layout()
plt.show()
```

`subplot()` 메서드의(x, y, index)의 인수를 xyindex로 붙여서 사용할 수 있음  
예시 : `subplot(2, 2, 1)` ▶ `subplot(221)`

```python
plt.subplot(221)
plt.plot(np.random.rand(5))
plt.title('첫번째 플롯')

plt.subplot(222)
plt.plot(np.random.rand(5))
plt.title('두번째 플롯')

plt.subplot(223)
plt.plot(np.random.rand(5))
plt.title('세번째 플롯')

plt.subplot(224)
plt.plot(np.random.rand(5))
plt.title('네번째 플롯')

plt.tight_layout()
plt.show()
```

`subplots(x, y)` 메서드로 복수의 Axes 객체를 담은 배열을 생성할 수 있음  
`subplots(x, y)` 메서드는 (figure, axes 배열) 튜플을 반환

```python
figure, axes = plt.subplots(2, 2)

axes[0, 0].plot(np.random.rand(5))
axes[0, 0].set_title('1행 1열')

axes[0, 1].plot(np.random.rand(5))
axes[0, 1].set_title('1행 2열')

axes[1, 0].plot(np.random.rand(5))
axes[1, 0].set_title('2행 1열')

axes[1, 1].plot(np.random.rand(5))
axes[1, 1].set_title('2행 2열')

plt.tight_layout()
plt.show()
```

```python
figure, axes0 = plt.subplots()
axes1 = axes0.twinx()

axes0.set_title('x축을 공유하는 플롯')
axes0.plot([10, 5, 2, 9, 7], 'r-', label='y0')
axes0.set_ylabel('y0')

axes1.plot([100, 200, 220, 180, 120], 'g:', label='y1')
axes1.set_ylabel('y1')

axes0.set_xlabel('x축')
plt.legend(loc=0)
plt.show()
```

```python
figure, axes0 = plt.subplots()
axes1 = axes0.twiny()

axes0.set_title('y축을 공유하는 플롯')
axes0.plot([10, 5, 2, 9, 7], 'r-', label='x0')
axes0.set_xlabel('x0')

axes1.plot([100, 200, 220, 180, 120], 'g:', label='x1')
axes1.set_xlabel('x1')

axes0.set_ylabel('y축')
plt.legend(loc=10)

plt.show()
```
