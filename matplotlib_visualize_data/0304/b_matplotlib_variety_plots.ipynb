#### 바 차트(막대 그래프)  
x 축의 값이 분류가 가능한 카테고리 형태일 때 `bar()`, `barh()` 메서드로 막대 그래프를 표시할 수 있음  
주의! bar 메서드는 y축의 값이 첫번째 인수로 옴

```python
y = [2, 3, 1]
x = [1, 2, 3]
xlabel = ['가', '나', '다']

plt.bar(x, y)
plt.xticks(x, xlabel)
plt.yticks(sorted(y))

plt.show()
```

##### 오차 막대
`xerr`, `yerr` 인수를 지정하면 에러 바(error bar)를 추가할 수 있음  
`alpha` 인수를 지정하면 바의 투명도를 지정할 수 있음(0: 완전투명, 1: 완전 불투명)  

```python
np.random.seed(0)

people = ['몽룡', '춘향', '방자', '향단']
y_position = np.arange(4)
y_performance = 3 + 10 * np.random.rand(4)
error = np.random.rand(4)

plt.title('Bar h')
plt.barh(y_position, y_performance, xerr=error, alpha=0.4)
plt.yticks(y_position, people)
plt.xlabel('x 축')
plt.grid(True)

plt.show()
```

#### 스템 플롯
이산확률이나 자기 상관관계를 표현하고자 할 때 사용하는 차트  
`stem()` 메서드로 표시할 수 있음

```python
x = np.linspace(0.1, 2*np.pi, 10)
plt.title('스템 플롯')
plt.stem(x, np.cos(x), '-.')
plt.grid(True)

plt.show()
```

#### 파이 차트  
각 분류가 전체에서 얼마만큼 차지하고 있는지 파악할 때 사용하는 차트  
`pie()` 메서드로 표현할 수 있음  
주의! 차트가 원형태를 유지하려면 `axis('equal')`을 지정해야 함

```python
labels = ['개구리', '돼지', '개', '고양이']
sizes = [15, 30, 45, 10]
colors = ['yellowgreen', 'gold', 'lightskyblue', '#ff00ff']
explode = (0, 0.1, 0, 0) # 떨어트리는거 표시

plt.title('파이 차트')
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=90)
plt.axis('equal')

plt.show()
```

#### 히스토그램  
특정 구간을 정해 그 구간에 대한 집계 데이터를 표시할 때 사용하는 차트  
`hist()` 메서드로 표시할 수 있음  
`bins` 인수로 집계할 구간의 정보를 받음

```python
x = np.random.randn(1000)

plt.title('히스토그램')
arrays, bins, patches = plt.hist(x, bins=10)
# arrays = array([  1.,   7.,  32.,  97., 184., 273., 229., 123.,  42.,  12.])
# bins = array([-3.69428529, -3.03648079, -2.37867629, ..., 2.8837597 ])
# patches = <BarContainer object of 10 artists>
plt.grid(True)

plt.show()
```

#### 스캐터 플롯  
두 변수 간의 상관관계를 나타내고자 할 때 사용하는 차트  
`scatter()` 메서드로 표현할 수 있음  
점 하나의 위치는 x, y값으로 표현됨

```python
x = np.random.normal(0, 1, 100)
y = np.random.normal(0, 1, 100)

plt.title('스캐터 플롯')
plt.scatter(x, y)
plt.grid(True)

plt.show()
```

만약 x에 대해서 여러 변수를 한 차트로 비교할 때 점의 크기 혹은 색상으로 표현할 수 있음  
`scatter()` 메서드의 `c` 인수로 색상을, `s`인수로 크기를 지정할 수 있음

```python
n = 30
x = np.random.rand(n)
y1 = np.random.rand(n)
y2 = np.random.rand(n)
y3 = np.pi * (15*np.random.rand(n)) **2

plt.title('버블 차트')
plt.scatter(x, y1, c=y2, s=y3)
plt.grid(True)

plt.show()
```

#### Imshow  
이미지데이터처럼 행과 열을 가진 2차원 데이터에 대해 각 행렬 값을 색으로 표현하는 차트  
`imshow()` 메서드로 표현할 수 있음 

```python
from sklearn.datasets import load_digits
```

```python
digits = load_digits()
x = digits.images[0]
x
```

```python
plt.title('mnist digis; 0')
plt.imshow(x, interpolation='nearest', cmap=plt.cm.bone_r) # 차트생성
# cmap 컬러변경
# interpolation 시각화표현
plt.subplots_adjust(left=0.35, right=0.65, bottom=0.35, top=0.65) # 크기 설정
plt.xticks([]) # 틱 가릴때 인수[] 지정
plt.yticks([]) # 틱 가릴때 인수[] 지정

plt.show
```

`cmap` 인수로 데이터를 표현하는 칼라맵을 지정할 수 있음  
`interpolation` 인수로 데이터의 시각화 표현을 지정할 수 있음

```python
figure, axes = plt.subplots(1, 4, figsize=(12, 3), subplot_kw={'xticks': [], 'yticks': []})

axes[0].set_title('color map Blues')
axes[0].imshow(x, interpolation='nearest', cmap=plt.cm.Blues)
axes[1].set_title('color map Blues r')
axes[1].imshow(x, interpolation='nearest', cmap=plt.cm.Blues_r)
axes[2].set_title('interpolation none')
axes[2].imshow(x, interpolation='none', cmap=plt.cm.bone_r)
axes[3].set_title('interpolation kaiser')
axes[3].imshow(x, interpolation='kaiser', cmap=plt.cm.bone_r)

plt.show()
```

#### 컨투어 플롯(등고선 차트)  
3차원 자료에서 x, y입력과 z출력이 존재할 때, z표현을 명암이 아닌 선을 사용하는 차트  
`contour()`, `contourf()` 메서드를 사용  
`contour()` : 등고선만 표시  // `contourf()` : 색으로 표시  
입력 x, y에 대하여 1차원 배열을 그대로 사용할 수 없고  
, `meshgrid()` 메서드로 포인트 행렬을 만들어 사용해야 한다

```python
def f(x, y):
  return (1 - x / 2 + x** 5 + y ** 3) * np.exp(-x ** 2 -y ** 2)

n = 256
x = np.linspace(-3, 3, n)
y = np.linspace(-3, 3, n)
xx, yy = np.meshgrid(x, y)
zz = f(xx, yy)

plt.title('컨투어 플롯')
plt.contour(xx, yy, zz, colors='black')
plt.contourf(xx, yy, zz, alpha=0.75, cmap='jet')
plt.grid(True)

plt.show()
```

#### 3D 서피스 플롯  
3차원 데이터를 입체적으로 표현하는 차트  
3D axes 객체를 생성하여 사용함  
`plot_wireframe()`, `plot_surface()` 메서드로 표현

```python
x = np.arange(-4, 4, 0.25)
y = np.arange(-4, 4, 0.25)
xx, yy = np.meshgrid(x, y)
rr = np.sqrt(xx**2 + yy**2)
zz = np.sin(rr)

axes = plt.axes(projection = '3d') # 3d 환경 만들기
axes.set_title('3d 서피스 플롯')
axes.plot_surface(xx, yy, zz, rstride=1, cstride=1, cmap='hot')

plt.show()
```
