```python
iris = seaborn.load_dataset('iris') # 붓꽃 데이터
titanic = seaborn.load_dataset('titanic') # 타이타닉 데이터
tips = seaborn.load_dataset('tips') # 팁 데이터
flights = seaborn.load_dataset('flights') # 여객기 데이터
```

#### seaborn 활용한 차트 겹치기
시각적 효과를 높이기 위해 여러 종류의 차트를 겹쳐서 표기 가능

```python
plt.title('박스 플롯, 스트립 플롯')
seaborn.boxplot(data=tips, x ='total_bill', y='day', whis=np.inf)
seaborn.stripplot(data=tips, x='total_bill', y='day')
plt.grid(True)

plt.show()
```

```python

plt.title('바이올린 플롯 + 스웜 플롯')
seaborn.violinplot(data=tips, x='day', y='total_bill')
seaborn.swarmplot(data=tips, x='day', y='total_bill')
plt.grid(True)
plt.show()
```

#### seaborn 활용한 스타일 변경  
seaborn 패키지는 matplotlib에서 제공하는 기본 스타일이 아닌 특정 스타일로 변경할 수 있도록 함  
`set()` 메서드로 색상, 틱 스타일 등을 변경할 수 있음  
`set_style()`  메서드로 틱 스 타일 변경할 수 있음

```python
def sinplot(flip=1):
  x = np.linspace(0, 14 , 100)
  for number in range (1, 7):
    plt.plot(x, np.sin(x + number * 0.5)* (7-number)*flip)
  plt.show()
```

```python
seaborn.set_style('ticks')
sinplot()
```

```python
seaborn.set_style('darkgrid')
sinplot()
```

```python
seaborn.set_style('dark')
sinplot()
```

```python
data = iris.sepal_length[:20]
data.plot.bar(rot=0)
plt.title('시리즈 바 차트')
plt.xlabel('Data')
plt.ylabel('꽃받침 길이')
plt.show()
```

```python
data = iris[:5]
data.plot.bar(rot=0)
plt.xlabel('Data')
plt.ylabel('Feature')
plt.ylim(0, 7)
plt.show()
```

```python
data = iris.groupby(iris.species).mean()
data.columns.name = 'feature' # 컬럼네임 설정시 범례에 표시 됨
data.plot.bar(rot=0)
plt.title('각 종의 feature 별 평균')
plt.xlabel('평균')
plt.ylabel('종')
plt.ylim(0, 9)
plt.grid(True)
plt.show()
```

```python
data = titanic.pclass.value_counts()
data.plot.pie(autopct = '%.2f%%')
plt.title('선실별 승객 수 비율')
plt.axis('equal')
plt.grid(True)
plt.show()
```

```python
iris.plot.hist()
plt.title('각 feature 값들의 빈도 수 히스토그램')
plt.xlabel('Data')
plt.grid(True)
plt.show()
```

```python
iris.plot.kde()
plt.title('각 feature 값들의 빈도 수 커널 분포 플롯')
plt.xlabel('Data')
plt.show()
```

```python
iris.plot.box()
plt.title('각 feature 값들의 반도 수 박스 플롯')
plt.xlabel('Feature')
plt.ylabel('Data')
plt.show()
```
